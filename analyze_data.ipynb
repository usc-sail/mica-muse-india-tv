```python
import math
import collections
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({"font.size": 18,
                     "font.family": "sans-serif",
                     "figure.figsize": (20, 8),
                     "axes.facecolor": "ffffff",
                     "figure.dpi"       : 200,
                     "legend.fontsize"  : "large",
                     "figure.titlesize" : "medium",
                     "lines.linewidth": 3,
})
```

```python
data_df = pd.read_csv("data_long_form.csv", index_col=None)
```

```python
data_df
```

```python
count_df = data_df.groupby(["video_key", "name", "lang", "year"])["faces"].sum().reset_index()
count_df = count_df.rename(columns={"faces": "total_faces"})
count_df = count_df.set_index(["video_key", "name", "lang", "year"])
count_df
```

```python
gender_df = data_df.groupby(["video_key", "name", "lang", "year", "gender"])["faces"].sum().reset_index()
gender_df = gender_df.set_index(["video_key", "name", "lang", "year"])
gender_df = gender_df.join(count_df)
gender_df["screentime"] = 100 * gender_df["faces"] / gender_df["total_faces"]
gender_df = gender_df.reset_index()
gender_df = gender_df[["video_key", "name", "lang", "year", "gender", "screentime"]]
gender_df
```
